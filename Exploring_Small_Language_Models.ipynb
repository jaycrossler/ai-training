```python
from transformers import pipeline

classifier = pipeline("sentiment-analysis")
classifier("I've been waiting for my whole life to join the Air Force.")
```

```python
classifier(
    ["I love 2 hour briefings. Not.", "My boss made me attend", "I for one welcome our new AI overlords."]
)
```

```python
from transformers import pipeline

classifier = pipeline("zero-shot-classification")
classifier(
    "This is an online course about AI for use on Air Force systems",
    candidate_labels=["education", "politics", "business", "technology", "military"],
)
```

```python
from transformers import pipeline

generator = pipeline("text-generation")
generator("In this Air Force course, we will teach you how to")
```

```python
from transformers import pipeline

generator = pipeline("text-generation", model="distilgpt2")
generator(
    "In this Air Force course, we will teach you how to",
    max_length=30,
    num_return_sequences=2,
)
```

```python
from transformers import pipeline

unmasker = pipeline("fill-mask")
unmasker("This course will teach you all about <mask> models and their use in Air Force mission systems.", top_k=2)
```

```python
from transformers import pipeline

ner = pipeline("ner", grouped_entities=True)
ner("My name is Jay and I work for The MITRE Corporation in the DC area. I frequently travel to Boston, New York, and LA.")
```

```python
from transformers import pipeline

question_answerer = pipeline("question-answering")
question_answerer(
    question="Where do I work?",
    context="My name is Jay and I work for The MITRE Corporation in the DC area. I frequently travel to Boston, New York, and LA."
)
```

```python
from transformers import pipeline

summarizer = pipeline("summarization")
summarizer(
    """
    Background

Our mission systems are increasingly more complex, and the time required to respond to world events is decreasing on average, while more tools are needed, and less budget is available. This leads to system solutions of higher complexity, faster delivery, and lower cost.  As evidenced by increasing compromises over time, software supply chains and Continuous Integration/Continuous Delivery (CI/CD) environments are attractive targets for malicious cyber actors.
Purpose and Methodology

This document captures the DevSecOps best practices commonly employed by government organizations for secure development performed in an unclassified environment.  Although this document focuses on secure development in the unclassified system; the tools, software, and resources under consideration should also operate securely in higher security domains.

This document presents a consensus of low-side software development best practices collected from government agencies.

Setting the context – Best Practices revolve around these Key Terms

There definitions will be used within the document, and based on feedback from the DevSecOps Community of Practice.

-       DevSecOps is an organization’s software engineering culture and practice that aims at unifying software development, security, and operations. The purpose of DevSecOps is to automate, monitor, and apply security at all phases of the software lifecycle (plan, develop, build, test, release, deliver, deploy, operate, and monitor). DevSecOps is the cornerstone of reducing risk in modern software development and deployment.  It relies on strong Configuration Management (CM) to trace and manage all software artifacts.
-       Continuous Integration (CI) involves developers frequently merging code changes into a central repository triggering automated builds and tests.  This represents a large process and cultural change where instead of a few large changes every few months, there are sometimes dozens of small changes per day to a system.  When done right, it reduces integration challenges and provides additional tracking of needs to work to functionality – which allows frequent updates and software more aligned to agile missions, along with business benefits of better tracking and predicting completion.
-       Continuous Monitoring (ConMon, sometimes also called CM not to be confused with Configuration Management’s CM) involves automating security and IT operations processes to consistently verify and check code, tools, apps, and configurations for vulnerabilities, misconfigurations, or configuration drift.  This approach supports frequent checks, such as "hourly" rather than "annually," which necessitates shifting to an active cyber defense strategy.  According to NIST SP 800-137, Information Security Continuous Monitoring (ISCM) for Federal Information Systems and Organizations, ISCM is defined as maintaining ongoing awareness of information security, vulnerabilities, and threats to support organizational risk management decisions.
-       Continuous Delivery (CD) is a software development practice where code changes are automatically prepared for a release to production or automatically deployed to test or beta infrastructures.  Upon a manual compliance approval step, releases can be automatically deployed into operational IT infrastructure.  This explicit approval step might be waived resulting in Continuous Deployment.
-       Software Supply Chain (SSC) is the combination of all the tools, libraries, and components used to develop, build, and publish software.  Sometimes adding one simple component can trigger the automatic inheritance of hundreds of other supporting components – a prime target for malicious cyber actors.
-       Software Bill of Materials (SBOM) is a key building block in software supply chain risk management. “An SBOM is a formal, machine-readable inventory of software components and dependencies, information about those components, and their hierarchical relationships.  These inventories should be comprehensive – and should explicitly state where they could not be.  SBOMs may include open source or proprietary software and can be widely available or access restricted.”
-       CI/CD pipelines and DevSecOps Platforms are the emerging sets of tools and techniques that automate, scan, and enable this modern system development.  These Platforms are massive force multipliers – though the more they are used, the greater targets they become.  These platforms should be embraced, while managing the risk.
Mission Systems: Why System Risk Cannot Be Zero

This document aims to help program offices and related stakeholders reduce risk by following tested best practices.
"""
)
```

```python
from transformers import pipeline

unmasker = pipeline("fill-mask", model="bert-base-uncased")
result = unmasker("This Air Force Technical Seargeant works as a [MASK]. He is tall.")
print([r["token_str"] for r in result])

result = unmasker("This Air Force Technical Seargeant works as a [MASK]. She is tall.")
print([r["token_str"] for r in result])
```
